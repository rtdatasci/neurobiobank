```python
from nilearn import plotting, datasets, image
import numpy as np

# Load the Harvard-Oxford Atlas
atlas = datasets.fetch_atlas_harvard_oxford('cort-maxprob-thr25-1mm')
atlas_filename = atlas.maps  # 3D atlas
labels = atlas.labels

# Find the index of the region
region_name = "Frontal Pole"  # Exact name from the labels
roi_index = labels.index(region_name)  # Find the index of the label

# Create a binary mask for the region
roi_img = image.math_img(f"img == {roi_index}", img=atlas_filename)

# Plot the binary mask
plotting.plot_roi(roi_img, title=f"Region: {region_name}", draw_cross=False)
plotting.show()
```
